Bar Graph: Top 10 Earning Degrees

```python
#Import and Setup

import pandas as pd
%matplotlib inline

#File
degreespay = "college-salaries/degrees-that-pay-back.csv"
df = pd.read_csv(degreespay)
df.head(3)
```

```python
#data frame to pare down on the information
major_salary = df[["Undergraduate Major", "Starting Median Salary"]]
major_salary.head()
```

```python
major_salary.head()
#Top 10 earning degrees 
sortdegrees = df.sort_values('Starting Median Salary', ascending=False)[:10]
sortdegrees.head(10)
```

```python
#Had to adjust the data in the CSV file to not include $ in starting median salary. 
#Creating graph, titles automatically created for X, I created the Y label
graph = sortdegrees.plot.bar(x = 'Undergraduate Major', y='Starting Median Salary', fontsize = 20, figsize = (15,10))
graph.set_xlabel("Undergraduate Degree", fontsize = 20)
graph.set_title("Top 10 Earning Degrees", fontsize = 20)
graph.set_ylabel("Salary in Dollars", fontsize = 20)
plt.show()
fig = graph.get_figure()
fig.savefig("Top10.png")
```
